# Functions

1. What is the difference between a function and a method in Python.

  - A function in Python is a block of code that performs a task and can be called independently. A method is a function associated with an object and is called using dot notation on the object.

2. Explain the concept of function arguments and parameters in Python.

  - In Python, parameters are variables defined in a function’s definition, while arguments are the values passed to the function when called. Parameters act as placeholders, and arguments provide the actual data. Functions can accept positional, keyword, or default arguments for flexibility.

3. What are the different ways to define and call a function in Python?

  - In Python, functions can be defined using the `def` keyword, followed by the function name and parameters. They are called by using the function name and passing the appropriate arguments. Functions can also be defined using lambda expressions.

4. What is the purpose of the `return` statement in a Python function?

  - The `return` statement in Python functions is used to exit the function and optionally pass a value back to the caller. It allows functions to produce results that can be used or stored elsewhere in the program.

5. What are iterators in Python and how do they differ from iterables?

  - In Python, an **iterator** is an object that implements the `__iter__()` and `__next__()` methods, allowing it to traverse through a sequence of values one at a time. An **iterable** is any object that can be iterated over, like lists or tuples, and implements the `__iter__()` method but does not necessarily implement `__next__()`. The key difference is that iterables can generate iterators, but iterators are the actual objects that iterate over the values.

6.  Explain the concept of generators in Python and how they are defined.

  - Generators in Python are special functions that use the `yield` keyword to produce values one at a time, pausing and resuming execution. They are defined like regular functions but return a generator object for iteration.

7. What are the advantages of using generators over regular functions?

  - Generators are more memory-efficient than regular functions because they yield values one at a time, without storing the entire sequence in memory. They also allow lazy evaluation, processing data only when needed, improving performance for large datasets.

8. What is a lambda function in Python and when is it typically used?

  - A lambda function in Python is a small, anonymous function defined using the `lambda` keyword. It can take any number of arguments but only has a single expression. It's typically used for short, simple functions in places like `map()` or `filter()`.

9. Explain the purpose and usage of the `map()` function in Python.

  - The `map()` function in Python applies a given function to all items in an iterable (like a list) and returns an iterator. It's useful for transforming data without using explicit loops, improving readability and efficiency.

10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

  -- In Python:

    - **`map()`** applies a function to each item in an iterable and returns an iterable of results.
    - **`reduce()`** (from `functools`) accumulates a result by applying a function to pairs of items, reducing the iterable to a single value.
    - **`filter()`** applies a function to each item in an iterable, keeping only those for which the function returns `True`.

```python
# 1.  Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    # Use a list comprehension to filter even numbers and sum them
    return sum(number for number in numbers if number % 2 == 0)

# Example usage
numbers = [47, 10, 42, 13, 8, 6]
result = sum_of_even_numbers(numbers)
print("Sum of even numbers:", result)
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(input_string):
    return input_string[::-1]

# Example usage
input_str = "Hello, World!"
reversed_str = reverse_string(input_str)
print("Reversed string:", reversed_str)
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

def square_numbers(numbers):
    return [number ** 2 for number in numbers]

# Example usage
numbers = [1, 2, 3, 4, 5]
squared_numbers = square_numbers(numbers)
print("Squared numbers:", squared_numbers)
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            return False
    return True

# Check primes between 1 and 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]

print("Prime numbers between 1 and 200:", prime_numbers)
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next_num = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            # Return the current Fibonacci number
            fib_number = self.current
            # Update for next number
            self.current, self.next_num = self.next_num, self.current + self.next_num
            self.count += 1
            return fib_number
        else:
            raise StopIteration

# Example usage:
fib_iterator = FibonacciIterator(10)  # Get first 10 Fibonacci numbers

for num in fib_iterator:
    print(num)
```

```python
# 6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
for power in powers_of_2(5):
    print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.

L = ["Geeks\n", "for\n", "Geeks\n"]

# Writing to file
file1 = open('myfile.txt', 'w')
file1.writelines(L)
file1.close()

# Opening file
file1 = open('myfile.txt', 'r')
count = 0

# Using for loop
print("Using for loop")
for line in file1:
    count += 1
    print("Line{}: {}".format(count, line.strip()))

# Closing files
file1.close()
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# List of tuples
tuples_list = [(1, 3), (2, 2), (4, 1), (3, 4)]

# Sorting the list of tuples based on the second element
sorted_list = sorted(tuples_list, key=lambda x: x[1])

# Display the sorted list
print(sorted_list)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 30, 40, 100]

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map() to apply the conversion function to each element in the list
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Display the result
print(fahrenheit_temperatures)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

# Function to check if a character is a vowel
def is_not_vowel(char):
    vowels = "aeiouAEIOU"
    return char not in vowels

# Input string
input_string = "Hello, World!"

# Use filter() to remove vowels from the string
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Display the result
print(filtered_string)
```

```python
# 11. Imagine an accounting routine used in a book shop. It works on a list with sublista, which look like this:

#OrderNumber,"Book Title and Author", Quantity,Price per Item

 #34587, "Learning Python, Mark Lutz", 4, 40.95),
 #98762, "Programming Python, Mark Lutz", 5, 56.80),
 #77226, "Head First Python, Paul Barry", 3, 32.95),
 #88112, "Einführung in Python3, Bernd Klein", 3, 24.99)

#Write a Python program, wish returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,- € if the value of the order is smalier than 100,00 €.

#Write a Python program using lambda and map.

# Input data: list of orders
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einführung in Python3, Bernd Klein", 3, 24.99)
]

# Function to calculate the total price and apply the additional €10 if necessary
def calculate_order(order):
    order_number, title, quantity, price_per_item = order
    total_price = quantity * price_per_item
    if total_price < 100:
        total_price += 10  # Increase by 10 if total is less than 100
    return (order_number, total_price)

# Use map() with a lambda function to process each order
result = list(map(lambda order: calculate_order(order), orders))

# Print the result
print(result)
```
